```python
# Calcula o seno de um angulo usando a série de Taylor com n = 3.

from math import sin, factorial

# Ângulo do teste.
x = 1.2

# Variável que receberá a soma.
soma = 0

# Número de somas.
n=3

# Loop que faz a soma n vezes.
for n in range(0,3):
  soma += (((-1)**n) * (x**(2*n+1))) / factorial(2*n+1)

# Exibe o resultado ao final da soma.
print("Sen(x) para x = 1.2 é %f" %soma)


"""
Usando sen(x) do módulo math,
para x = 1.2, sen(x) = 0.932039,
já no algorítmo acima, sen(x) = 0.932736.
"""
```
