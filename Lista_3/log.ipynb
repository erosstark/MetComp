```python
from math import log

x = 1.2
n=4
soma = 0
for n in range(1,n):
  soma += 2 * ((1 / (2 * n - 1)) * ((x - 1) / (x + 1)) ** (2 * n - 1))
print(f"O log(x) para x = 1.2 é {soma}" )

print(log(x))

"""
Usando log(x) do módulo math o resultado
para x=1.2 é 0.1823215567939546.
Usando a série com n = 4, obtemos 0.18232154203740839.
Conclui-se que n=4 é uma ótima aproximação.
"""
```
