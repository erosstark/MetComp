```python
# Calcula o cosseno de um angulo usando a série de Taylor com n = 4.

from math import  factorial, cos, pi

# Lista com ângulos sugeridos pelo exercício.
angles = [pi/6, pi/3]

# Número de somas.
n = 4

# Lista vazia que receberá os valores do cosseno de cada ângulo.
cosx = []

# Variável que terá a soma atribuida a ela para ser acrescentada na lista.
cosxi = 0

# Loop que atribui a x os valores da lista "angles".
for x in angles:

  # Esse loop realiza a soma n vezes com o valor de x atribuido anteriormente.
  for n in range(0,n):
    cosxi += ((-1)**n) * (x**(2*n)) / factorial(2*n)

  # Acrescentando a soma à lista.
  cosx.append(cosxi)

  # Reseta a variável para receber a próxima soma a ser acrescentada na lista.
  cosxi = 0

# Exibe os resultados.
print("Cos(x) para x = pi/6 é %f" %(float(cosx[0])))
print("Cos(x) para x = pi/3 é %f" %(float(cosx[1])))

"""
 Usando o cos(x) do módulo math, os valores foram
x = pi/6, cos(x) = 0.866025 e
x = pi/3, cos(x) = 0.500000. Comparando com os valores da série
de Taylor com n = 4, conclui-se que é uma ótima aproximação.
"""
```
