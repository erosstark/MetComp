```python
#calcula o valor da funçâo Gaussiana dados valores de m, s e x.

import math

#pergunta os valores das variáveis.
m = float(input("qual valor de m? "))
s = float(input("qual valor de s? "))
x = float(input("qual valor de x? "))

#calcula a funçao gaussiana
f = (1 / (s * math.sqrt(2 * math.pi))) * math.exp(-(1 / 2) * ((x - m) / s) ** 2)

print(f)
```
