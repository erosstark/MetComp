```python
#Programa que converte coordenadas cartesianas para coordenadas polares

import math

#pergunta as coordenadas
x = float(input("coordenada x "))
y = float(input("coordenada y "))

#convers√¢o
r = round(math.hypot(x,y),2)
theta = (math.atan2(y,x))

#exibe o resultado em radianos
print(f"A coordenada cartesiana ({x},{y}) em coordenada polar fica: \nradianos \
({r}, {round(theta,2)})")


#exibe o resultado em graus
print(f"graus ({r}, {round(math.degrees(theta),2)})")


```
