```python
# O programa calcula a área de um poligono de N lados de comprimento L.

import math

# Entrada dos valores das variáveis.
n = int(input("quantos lados tem o poligono? "))
l = float(input("qual o comprimento de cada lado? "))

# Calcula a área.
a = (n * l ** 2) / (4 * math.tan(math.pi / n))

print(f"a área do poligono é {round(a,2)}m^2")
```
