```python
# Calcula o volume de um cilindro e a incerteza.

import math

#pergunta as variáveis.
r = float(input("qual o raio do cilindro? "))
h = float(input("qual a altura do cilindro? "))
ir = float(input("qual a incerteza do raio? "))
ih = float(input("qual a incerteza da altura? "))

# Calcula o volume.
v = float(math.pi * h * r**2)

"""
 A incerteza do volume é calculada usando o metodo
de propagação de incerteza de uma funçao geral.
"""
iv = float(math.hypot(2 * math.pi * r * h * ir, math.pi * (r**2) * ih))

print(f"o volume do cilindro e sua incerteza é: ({v} +ou-{iv})m^3")

"""
 Não encontrei uma maneira de ajustar as casas
decimais de acordo com a incerteza.

"""
```
