```python
# Programa calcula a distância entre dois pontos na superfície da terra usando
#latitude e longitude dadas em grau.

import math

# Entrada das variáveis em graus.
print(" A partir da latitude e longitude de dois pontos na terra, \
é possivel calcular a distância entre eles.\
\n A distância entre: \
\nSão Paulo, SP      23°32'56.0''S 46°38'20.0''W \
\nRio de Janeiro, RJ 22°54'13.0''S 43°12'35.0''W  \
\né 357.66km. O valor encontrado pela calculadora do IMPE foi 356546.132 m \
\n Digite a latitude e longitude em graus, minutos e segundos.\
\nSe atente ao sinais: Sul e Oeste(-), Norte e leste(+)")

print("\n Latitude ponto 1")
t1 = float(input("Graus° " )), float(input("Minutos ' ")), float(input('Segundos " '))
print("\n Longitude ponto 1")
g1 = float(input("Graus° " )), float(input("Minutos ' ")), float(input('Segundos " '))

print("\n Latitude ponto 2")
t2 = float(input("Graus° " )), float(input("Minutos ' ")), float(input('Segundos " '))
print("\n Longitude ponto 2")
g2 = float(input("Graus° " )), float(input("Minutos ' ")), float(input('Segundos " '))

# Conversão para radianos.

t1rad = math.radians(t1[0] + (t1[1]/60) + (t1[2]/3600))
g1rad = math.radians(g1[0] + (g1[1]/60) + (g1[2]/3600))

t2rad = math.radians(t2[0] + (t2[1]/60) + (t2[2]/3600))
g2rad = math.radians(g2[0] + (g2[1]/60) + (g2[2]/3600))



# Calculo da distância
r = float(6371.01)        # Raio da terra
d = r * math.acos(math.sin(t1rad) * math.sin(t2rad) \
+ math.cos(t1rad) * math.cos(t2rad) * math.cos(g1rad - g2rad) )

print(f"A distância entre os pontos fornecidos é {d:.2f}km")
```
