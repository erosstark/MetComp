```python
import numpy as np
import matplotlib.pyplot as plt

valores = np.loadtxt("/content/drive/MyDrive/Colab Notebooks/Lista 5/densidade_agua.txt")

temperatura = valores[:,0]

densidade = valores[:,1]

erro = valores[:,2]


plt.errorbar(temperatura, densidade,yerr= erro)

plt.xlabel("Temperatura (°C)")
plt.ylabel("Densidade (kg/$m^3$)")
plt.title("Densidade da água em função da temperatura")
plt.savefig("plot_density.png")
plt.show()
```
