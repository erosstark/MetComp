```python
import numpy as np
import matplotlib.pyplot as plt

gravidade = np.loadtxt("/content/drive/MyDrive/Colab Notebooks/Lista 5/gravidade.txt")

N = np.size(gravidade)
valor_min, valor_max = np.min(gravidade), np.max(gravidade)


intervalo = {min:975, max:995}

m = 60
delx = (intervalo[max] - intervalo[min]) / 60

Ni = np.array([])
gi = np.array([])
i = 0
contador = 0
while i < m:
  gi = np.append(gi,((intervalo[min]+ (delx * i)-(delx/2))))
  contador = 0
  for v in gravidade:
    if intervalo[min] + (delx * i) < v <= intervalo[min] + (delx * (i + 1)):
      contador += 1
  Ni = np.append(Ni,contador)

  i += 1


medio = np.mean(gravidade)
desv_padrao = np.std(gravidade, ddof = 1)
plt.bar(gi, Ni, width = delx, edgecolor="b")
plt.xlabel("g(cm/$s^2$)")
plt.ylabel("Contagem em cada intervalo")
plt.figtext(0.15,0.78,f"Valor médio $\sigma$={medio:.2f}cm/$s^2$ \
\nDesvio padrão $\mu$={desv_padrao:.2f}",backgroundcolor = (0, 0.5, 1, 0.3))



```
