```python
import numpy as np
import matplotlib.pyplot as plt

gravidade = np.loadtxt(r"C:\Users\eross\OneDrive\Documents\vscode python\MetComp\Lista_5\gravidade.txt")

min = 975 ; max = 995
bins = 20
delx = (max - min)/bins







n = 0
soma = 0
while n < bins:
  print (n)
  contador = 0
  for v in gravidade:
    if min + (delx * (n)) <= v < min + (delx * (n+1)):
      contador += 1
  plt.bar(n,contador,color = "r")
  print("contador %d" %contador)
  n += 1
  soma += contador 

plt.show()
```
