```python
import numpy as np
import matplotlib.pyplot as plt

 #np.arange(start,stop,step)
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)


plt.plot(x, y, 'ro') # r de red; o ´e um c´ırculo
plt.xlabel('angulo (0 a 4 pi)')
plt.ylabel('sen(x)')
plt.title('Grafico de seno(x)')
plt.axhline(color='gray', y=0, zorder=-1)
```
