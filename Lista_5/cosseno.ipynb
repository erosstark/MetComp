```python
import numpy as np
from math import factorial
import matplotlib.pyplot as plt

def taylor_cosseno(x,N):
  """
  Retorna uma aproximação para cosseno(x) usando a série de Taylor.
  x deve ser dado em radianos.
  N é o expoente do termo até onde a série será calculada.

  Ex:
  X = np.pi/6
  N = 6
  y = taylor_cosseno(x,N)

  A série será calculada até o termo com expoente 6.

  """
  cosx = 0
  for n in range(0,(N//2)+1):
    cosx += (((-1) ** n) * (x **(2 * n ))) / factorial(2 * n)
  return cosx



# Para x = pi/6 e 3pi/4.
Ns = [4,6]
xs = [np.pi/6, 3*np.pi/4]
xstr = ["pi/6","3pi/4"]        # Texto para ser exibido.
i = 0                          # index da lista do texto a ser exibido.

for x in xs:

  for N in Ns:

    # Aproximação usando a série de Taylor.
    cos_aprox = taylor_cosseno(x,N)
    print("N = %d  x = %s \nValor aproximado do cosseno(%s) é: %f" %(N, xstr[i], xstr[i], cos_aprox))

    # Diferença entre a aproximação e o valor obtido com np.cos(x).
    print("Diferença entre a aproximação e np.cos(%s) = %f " %(xstr[i], abs(np.cos(x) - cos_aprox)))

    # Proximo termo da série.
    prox_termo = taylor_cosseno(x,N+2) - taylor_cosseno(x,N)
    print("Valor do próximo termo da série : %f \n " %( prox_termo))

  i += 1



# Valores de x para plotar o gráfico

x = np.linspace(-np.pi/12, np.pi, 51)

# Gráfico np.cos(x)
plt.plot(x, np.cos(x),"r")

# Gráfico N=2
N=2
y2 = taylor_cosseno(x,N)
plt.plot(x,y2,"g")

# Gráfico N=4
N=4
y4 = taylor_cosseno(x,N)
plt.plot(x,y4)

# Gráfico N=6
N=6
y6 = taylor_cosseno(x,N)
plt.plot(x,y6,"b")

# Formatação do gráfico
plt.ylim(top = 2, bottom = -2)
plt.xticks(np.linspace(0, np.pi + np.pi/9 , 4, endpoint =False),['0','50','100','150'])
plt.title("Série de Taylor - Cos(x)")
plt.xlabel("x(graus)")
plt.ylabel("Cos(x)")
plt.legend(["Cos(x)","N=2" , "N=4", "N=6"])
plt.grid(True)


plt.savefig("cosseno.png")
plt.show()
```
