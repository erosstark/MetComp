```python
# Modelo rudimentar.


import numpy as np
import matplotlib.pyplot as plt


# Função a ser integrada.
f = lambda x: np.sin(x)

# Gráfico.
x = np.linspace(0,2*np.pi,50)
plt.plot(x,f(x))
plt.xlabel("x")
plt.grid(True)
plt.axhline(color = "red", zorder = 1)
plt.show()


#--------------------------------------------------
def trapezio(f, a, b, intmax = 100, err = 10**(-6)):
    t_anterior = 10**16
    t = 0.5 * (f(a) + f(b)) # calcula os extremos da função.
    for k in range(0,intmax):
        DELX_k = (b-a) / 2**k
        fj = 0
        for j in range(1,2**k):
            fj += f(a + j*DELX_k)

        tk  = (DELX_k) * (fj + t)

        if abs((tk-t_anterior)/tk) < err:
            break
        else:
            t_anterior = tk
    return tk,k
#------------------------------------------------------

# Variáveis.

a, b = 0, np.pi
print(trapezio(f,a,b))




```

```python
import numpy as np
def trapezia(fun, a, b, k):
    intmax = 100
    e = 10**(-6)
    tk_anterior = 10**16
    tk = 0.5 * (fun(a) + fun(b))  # Adicionando os valores nos extremos
    while k < intmax:
        h = (b - a) / (2**k)  # Calculando o tamanho de cada subintervalo


    # Somando os valores dentro dos subintervalos
        for j in range(1, 2**k):
            tk += fun(a + j * h)

        tk *= h  # Multiplicando pela largura do intervalo

        if abs((tk-tk_anterior)/tk) < e:
            intmax = k
        else:
            tk_anterior = tk
            k += 1
    return tk, k

# Exemplo de uso
exemplo_funcao = lambda x: np.sin(x)

a = 0  # Limite inferior
b = np.pi # Limite superior
k = 0  # Índice de T na fórmula


resultado = trapezia(exemplo_funcao, a, b, k)
print("Resultado da integral:", resultado)
```
