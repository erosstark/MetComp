```python
#   Calcula a integral sin(x) 0->pi usando o método de Simpsons em uma função
# em um loop ate atingir uma precisão 10^-6 e compara com o valor
# esperado conhecido.

# !!!!!!  Fórmula do método de Simposon nesse programa nao está em função
# do método dos trapézios.

import numpy as np



#-------------------------------------------------------------
def simpson(fun,a,b,k):
    """
        Calcula a integral de uma função
    usando o método de Simpsons.

    Variáveis:
        fun - função a ser integrada.
        a, b - limites de integração.
        k - índice.

        Retorna a integral aproximada até
    o índice k.

    """
    delx = (b-a)/2**k
    s = fun(a) + fun(b)
    fj = 0
    fi = 0
    for j in range(1,2**k,2):
        fj += fun(a + j*delx)
        i = j + 1
        fi += fun(a + i*delx)
    sk = (delx/3) * (4*fj + 2*fi + s)

    return sk
#----------------------------------------------------------------

# Função para ser integrada.
f = lambda x: np.sin(x)

# Limites de integração.
a, b =0, np.pi

# Número máximo de iteração e precisão.
intmax = 100
e = 10**(-6)

# Loop que chama a função simpson e vérifica se a precisão
# desejada foi alcançada.
print(f"O valor esperada para a integral de sin(x) 0->pi é 2. \n\
Aproximando com o método de Simpsons até a precisão relativa {e} \n")
print(" k   Aproximação            Diferença relativa para o valor esperado")

s_anterior = 10**16
k = 1
while k < intmax:
    s = simpson(f,a,b,k)
    print(f"{k:2d} | {s:10.10e}    |  {abs(2 - s)/2:.10e}")

    # Verifica a precisão.
    if abs((s - s_anterior) / s) < e:
        break
    else:
        s_anterior = s
        k += 1




```
