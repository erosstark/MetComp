```python
""" 
Eros Moreira Ferreira
24/04/2024

Calcula o coeficiente angular e linear de uma reta
dado dois pontos.
"""
import numpy as np

def inclinacao(x1, y1, x2, y2)->float:
  global a
  a = (y2-y1) / (x2-x1)
  return a
def intersecta(x1, y1)->float:
  
  return  y1-a*x1


x1,y1 = float(input("x1")), float(input("y1"))
x2,y2 = float(input("x2")), float(input("y2"))

print("coeficiente angular",inclinacao(x1, y1, x2, y2))
print("coeficiente linear" ,intersecta(x1, y1))

```
