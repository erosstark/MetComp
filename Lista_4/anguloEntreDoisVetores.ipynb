```python
import numpy as np

def modulo(r):
  soma = 0
  for i in r:
    i = i**2
    soma += i
    modulo = np.sqrt(soma)
  return modulo

def escalar(r, s):
  soma_vetor = r+s
  soma = 0
  for i in soma_vetor:
    soma += i
  return soma

def angulo(r,s):

  modulo_r = modulo(r)
  modulo_s = modulo(s)
  escalar_rs = escalar(r,s)

  cosx = escalar_rs / (modulo_r * modulo_s)
  x = np.arccos(cosx)
  return x


vetor1 = np.array([1,3,2])
vetor2 = np.array([2,3,2])

print(modulo(vetor1))
print(escalar(vetor1,vetor2))
print(angulo(vetor1,vetor2))

"""
precisa incrementar o input do teclado para os vetores
e testar direito
"""

```
