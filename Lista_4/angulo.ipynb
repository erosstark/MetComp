```python
import numpy as np

def angulo(x,y):
  """
   Calcula o ângulo entre dois vetores x e y de n dimensões
  """
  prod_escalar = np.dot(x,y)
  prod_modulo = np.linalg.norm(x) * np.linalg.norm(y)
  cos_angulo = (prod_escalar / prod_modulo)
  return np.arccos(cos_angulo)

x = np.array([1,1], int)
y = np.array([-1,-1], int)

print(angulo(x,y))
```
