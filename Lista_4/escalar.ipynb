```python
from numpy import array, dot

def escalar(x,y):
  """
  Calcula e retorna o produto escalar entre
  dois arrays x e y.
  """
  prod = x*y
  prod_escalar = 0
  for i in prod:
    prod_escalar += i
  return prod_escalar


x = array([3,4,5], float)
y = array([3,2,1], float)
print(escalar(x,y))

# Usando função dot do numpy
print(dot(x,y))
```
