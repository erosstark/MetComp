```python

from math import pi

# função que calcula o volume do cilindro.
def volCilindro(h,r):
  """
    Calcula e retorna o volume de um cilindro
    de altura h e raio r.
  """
  volume = pi * h * r ** 2
  return volume

# input da variáveis.
r = float(input("Raio r do cilindro "))
h = float(input("Altura h do cilindro "))

# chama a funçaão e exibe o resultado
print("O volume do cilindro é %f" %volCilindro(h,r))
```
