```python
def horas(tempo):
  horas = tempo // 3600
  return horas

def minutos(tempo):
  minutos = (tempo % 3600) // 60
  return minutos

def segundos(tempo):
  segundos =  (tempo % 3600) % 60
  return segundos



tempo = 3600*3+60*23+55
print("%d horas" %horas(tempo))
print("%d minutos" %minutos(tempo))
print("%d segundos" %segundos(tempo))
```
