```python


i = 1
j = 2
k = 3
n = int(input("n"))
contador = 0
while contador <= n/2:
  if i*j*k == n:                         """ nao consegui achar uma forma mais elegante"""
    print(i,j,k)
    contador = n
  elif contador > n/3:
    print("nao e")
    contador +=1
  else:
    contador += 1
    i += 1
    j += 1
    k += 1
```
