```python
# Diz se um dado número n inteiro positivo é primo ou não.

n = int(input("Digite um número inteiro positivo "))

i=n-1

# Calcula o maior divisor de n diferente dele mesmo.
if i == 0:
  print('%d é primo' %n)
else:
  while n % i != 0:
    i= i-1

  # Verifica se n é primo.
  if i > 1:
     print("%d não é primo" %n)
  else:
     print('%d é primo' %n)
```
