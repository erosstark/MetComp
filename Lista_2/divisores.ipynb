```python
# Calcula todos os divisores de um dado número n 
# inteiro positivo e a quantidade.

n = int(input("Digite um número inteiro positivo "))

if n > 0:
i = 1
numero_divisores = 0
print ("Divisores:")
while i <= n:
  if n % i == 0:
    print("%12d" % i)
    numero_divisores += 1
    i += 1
  else:
    i += 1

print("Total de divisores:",numero_divisores)



```
