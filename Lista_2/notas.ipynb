```python
# Calcula a quantidade m´ınima de notas que um banco deve fornecer para atingir
# o valor, sabendo que as notas disponíveis são de 100, 50, 20, 10 e 5 reais.

n = int(input("Valor inteiro"))
soma = 0
contador = 0
while soma < n:
  if soma + 100 <= n:
    soma += 100
    contador += 1
    print("100")
  elif soma + 50 <= n:
    soma += 50
    contador += 1
    print("50")
  elif soma + 20 <= n:
    soma += 20
    contador += 1
    print("20")
  elif soma + 10 <= n:
    soma += 10
    contador += 1
    print("10")
  else:
    soma += 5
    contador += 1
    print("5")

print("A quantidade mínima de notas que o banco deve fornecer para atingir o valor é:",contador)




```
