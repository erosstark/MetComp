```python
# Diz se um ponte de cooerdenada xy pertence a uma regiâo.

print("A região H é a parte situada no primeiro quadrante do plano xy de um círculo de raio 1 ")

# Entrada das variáveis
x = input("cooerdenada x positiva: ")
y = input("cooerdenada y positiva: ")

# Verificação
if float(x)<=1 and float(y)<=1:
  print(f"os pontos {x} e {y} pertencem a região H")
else:
  print(f"os pontos {x} e {y} não pertencem a região H")
```
