```python
# Encontra as raizes de uma equação de segundo grau.

import math
import cmath

print(""" Uma equação de segundo grau tem o formato ax^2 + bx + c = 0
e possui duas raizes que podem ser reais ou complexas.
""")

# Entrada das variáveis a,b e c.
a = float(input("digite o valor de a "))
if a != 0:
  b = float(input("digite o valor de b "))
  c = float(input("digite o valor de c "))


# Cáculo das raizes.
  delt = (b**2) - 4 * a * c

  # Se as raizes forem reais e distintas.
  if delt > 0:
    x1 = (math.sqrt(delt) - b) / 2 * a
    x2 = (-math.sqrt(delt) - b) / 2 * a
    print("A equação %5.2f x^2 + %5.2f x + %5.2f possui:" %(a,b,c))
    print("duas raizes reais distintas \n x1 = %.2f \n x2 = %.2f" %(x1,x2))

  # Se as raizes forem reais e iguais.
  elif delt == 0:
    x1 = (math.sqrt(delt) - b) / 2 * a
    x2 = (-math.sqrt(delt) - b) / 2 * a
    print("A equação %5.2f x^2 + %5.2f x + %5.2f possui:" %(a,b,c))
    print("duas raizes reais iguais \n x1 = %.2f \n x2 = %.2f" %(x1,x2))

  # Se as raizes forem complexas.
  else:
    x = (cmath.sqrt(delt) - b) / 2 * a
    print("A equação %5.2f x^2 + %5.2f x + %5.2f possui:" %(a,b,c))
    print("duas raizes complexas")
    re = x.real
    im = x.imag
    print("%5.2f +- %5.2f i "%(re,im))

# Se "a" for zero.
else:
  print(' "a" preicsa ser um número real diferente de zero.')

```
