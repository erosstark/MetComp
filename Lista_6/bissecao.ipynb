```python
import numpy as np
import matplotlib.pyplot as plt


#-----------------------------------------------------------------
# Função de x que téra a raiz calculada.
f = lambda x: (np.exp(-x) - np.sin((np.pi * x) / 2))

#-----------------------------------------------------------------


#-----------------------------------------------------------------
# Gráfico da função.
x = np.linspace(0,4,100)
plt.plot(x,f(x))
plt.xlabel("eixo X")
plt.ylabel("f(x)")
plt.grid(True)
plt.axhline(color = "r", zorder = 1)
plt.show()
#-----------------------------------------------------------------


#-----------------------------------------------------------------
def raiz(xmax,xmin):
  """
    Define a raiz de uma função f(x) usando o
  metodo da bisseção.

    É necessário informar o intervalo no qual deseja
  saber a raiz. O metodo sempre converge para UMA raiz caso ela
  exista e esteja no intervalo inicial.

    A precisão é de 10^-7 e o número de iterações e 100

  LIMITAÇÕES!!!
    O método não serve para:
      intervalos que possuem mais de uma raiz.
      Intervalos com singularidade.
      Raiz que não cruza o eixo x.

  Retorna uma tuple com a raiz e o número de iterações.
  """
  precisao = 10 ** (-7)
  nit = 0
  limite = 100
  fmin = f(xmin)

  # Verifica se o intervalo é valido.
  if f(xmax) * fmin > 0:
    print("intervalo inadequado")
  else:
    # Loop que calcula a raiz.
    while abs(xmax-xmin) > precisao and nit < limite:
      xmedio = (xmax+xmin) / 2
      fmedio = f(xmedio)
      if fmin * fmedio < 0:
        xmax = xmedio
      else:
        xmin = xmedio
      nit += 1
    return xmedio, nit
#--------------------------------------------------------------


print(" Observe o gráfico acima e digite um intervalo apropriado para ter \
a raíz caculada. \n O intervalo não pode conter mais de uma raíz.")
# Intervalo da raiz.
x1 = float(input("Intervalo mínimo: "))
x2 = float(input("intervalo máximo: "))


"""
resolve essa porra do return ai depois fdp.
"""
# Atribui os valores da tuple a função retorna às respectivas variáveis.

r, n_iteraçoes = raiz(x1,x2)


print(f"O numero de iterações foi {n_iteraçoes} e a raiz é {r:.7f}")



```
